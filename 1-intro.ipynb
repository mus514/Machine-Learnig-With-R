## Bias variance tradoff

```r
f <- function(X, n = 1000)
{
    3 * X[,1] + 0.4 * X[,2] + rnorm(n, 0, 4)
}
```

```r
get_sim_data <- function(rho, n = 1000)
{
    mu <- c(0, 0)
    Sigma <- matrix(c(1, rho, rho, 1),ncol = 2)
    X <- MASS::mvrnorm(n = n, mu = mu, Sigma = Sigma)
    Y <- f(X)
    data.frame(x1 = X[,1], x2 = X[,2], y = Y)
}
```

## Likelihood

```r
y <- rnorm(10, 2, 1)
```

```r
f <- function(theta)
{
    sum(dnorm(y, theta, 1, log = TRUE))
}
```

```r
x <- seq(-4, 4, by = 0.01)
```

```r
plot(x, Vectorize(f)(x))
```

```r
y <- rgamma(10, 0.5, 2)
```

```r
f <- function(theta)
{
    prod(dgamma(y, theta, 2))
}
```

```r
x <- seq(0, 3, by = 0.001)
```

```r
plot(x, Vectorize(f)(x))
```
